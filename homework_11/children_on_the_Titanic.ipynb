## 4. Количество выживших взрослых и количество выживших детей на Титанике:

Узнаем общее число выживших людей на Титанике:

```python
# Выбор определенного кол-ва рядов и столбцов
# используется метод loc в [], первый аргумент индекс или селектор, а второй список со столбцами
df.loc[df['Survived'] == 1, ['Name', 'Age']]
```

Узнаем сколько было людей старше 18 лет и сколько из них выжило:

```python
# Выбор определенного кол-ва рядов
# Синтаксис df[df[col] !=|==|>|<| значение]
df.loc[df['Age'] > 18, ['Name','Age']]
```

```python
df.loc[((df['Age'] > 18) & (df['Survived'] == 1)), ['Name','Age']]
```

Узнаем какое количество детей до 18 лет было на борту Титаника:

```python
df.loc[df['Age'] <= 18, ['Name','Age']]
```

Узнаем какое количество детей выжило на Титанике:

```python
# Для отбора можно использовать несколько условий одновременно
# Знак & означает 'and', а знак | 'or'
df.loc[((df['Age'] <= 18) & (df['Survived'] == 1)), ['Name','Age']]
```

* Из 887 человек, плывших на Титанике, выжило 342 человека.
* Людей старше 18 лет на Титанике было 721 человек, из них выжило 264 человека. Что составляет 36%.
* Из 166 детей и подростков до 18 лет выжило 78 детей. Что составляет 47%.

## 7. Зависимость количества выживших детей от наличия родственников на борту:
